Fix missing values in df_orders:

```python
#check how many rows the dataframe has in total
df_orders.describe()
```

The big majority of the empty cells in order_approved_at, order_delivered_carrier_date and order_approved_at come from the same rows. In total, less than 5% missing values were found. So they do not have further impact and will be deleted. 

```python
# replacing rows with empty values with NAN in df_orders to delete them in the next step
df_orders['order_approved_at'].replace('', np.nan, inplace=True)
df_orders['order_delivered_carrier_date'].replace('', np.nan, inplace=True)
df_orders['order_delivered_customer_date'].replace('', np.nan, inplace=True)
```

```python
# deleting the rows with missing values in df_orders
df_orders.dropna(subset=['order_delivered_customer_date'], inplace=True)
df_orders.dropna(subset=['order_delivered_carrier_date'], inplace=True)
df_orders.dropna(subset=['order_approved_at'], inplace=True)
```
